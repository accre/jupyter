```r
# Shell magic commands/cells are not supported in R kernels so we rely on the R system() command for 
# calls to the Unix shell instead
rversion=system('R --version| head -1 | cut -f3 -d " "',intern=TRUE) # Get version of R
mklocaldir=paste('mkdir -v -p ~/R/rlib-',rversion, sep="") # Generate string to pass to next line
system(mklocaldir,intern=TRUE) # Make a local directory to install R packages to
libdir=paste('~/R/rlib-',rversion, sep="") # Generate string to pass to next line
.libPaths(libdir) # Add local directory to our R search path
```

```r
install.packages("ggplot2") # Install ggplot2 to our local directory (if it's not already there)
```

```r
# Tutorial taken from: http://r-statistics.co/Complete-Ggplot2-Tutorial-Part1-With-R-Code.html
# Setup
options(scipen=999)  # turn off scientific notation like 1e+06
library(ggplot2)
data("midwest", package = "ggplot2")  # load the data
# midwest <- read.csv("http://goo.gl/G1K41K") # alt source 

# Init Ggplot
ggplot(midwest, aes(x=area, y=poptotal))  # area and poptotal are columns in 'midwest'
```

```r
ggplot(midwest, aes(x=area, y=poptotal)) + geom_point()
```

```r
g <- ggplot(midwest, aes(x=area, y=poptotal)) + geom_point() + geom_smooth(method="lm")  # set se=FALSE to turnoff confidence bands
plot(g)
```

```r
g <- ggplot(midwest, aes(x=area, y=poptotal)) + geom_point() + geom_smooth(method="lm")  # set se=FALSE to turnoff confidence bands

# Delete the points outside the limits
g + xlim(c(0, 0.1)) + ylim(c(0, 1000000))   # deletes points
# g + xlim(0, 0.1) + ylim(0, 1000000)   # deletes points
```

```r
g <- ggplot(midwest, aes(x=area, y=poptotal)) + geom_point() + geom_smooth(method="lm")  # set se=FALSE to turnoff confidence bands

# Zoom in without deleting the points outside the limits. 
# As a result, the line of best fit is the same as the original plot.
g1 <- g + coord_cartesian(xlim=c(0,0.1), ylim=c(0, 1000000))  # zooms in
plot(g1)
```

```r
g <- ggplot(midwest, aes(x=area, y=poptotal)) + geom_point() + geom_smooth(method="lm")  # set se=FALSE to turnoff confidence bands

g1 <- g + coord_cartesian(xlim=c(0,0.1), ylim=c(0, 1000000))  # zooms in

# Add Title and Labels
g1 + labs(title="Area Vs Population", subtitle="From midwest dataset", y="Population", x="Area", caption="Midwest Demographics")

# or

g1 + ggtitle("Area Vs Population", subtitle="From midwest dataset") + xlab("Area") + ylab("Population")
```
